# Data Visualization Practice on a Dummy Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head(10)
```

# bar and Countplot
IT helps you to aggregate data of a feature in your data.
Barplot is a general plot that allow ypou to aggregate the categorical data based off some function to default the mean.

```python
sns.barplot(x = 'sex', y = 'total_bill', data = tips);
#when you dont specify the estimator it takes the mean
```

```python
group = tips.groupby('sex')
group.total_bill.mean()
```

```python
sns.barplot(x = 'sex', y = 'total_bill', data = tips, estimator = np.std);
```

# Count plot
This only counts the number of occurences using the bar graph 

```python
sns.countplot(x = 'smoker', data = tips);
```

```python
sns.countplot(x = 'sex', data = tips, hue = 'smoker', palette = 'rocket');
#hue further separates your dataset into categories
```

# Box and Violin plots
Violin is use to show the distribution of categorical data

```python
plt.figure(figsize=(10,4))
sns.boxplot(x = 'day', y = 'total_bill', data = tips, palette = 'rainbow', hue = 'sex');
```

```python
plt.figure(figsize=(5,4))
sns.boxplot(y = 'day', x = 'total_bill', data = tips, palette = 'CMRmap', hue = 'sex');
```

```python
sns.boxplot(data = tips, palette = 'CMRmap', orient = 'h');
#orient can either be h or v
```

# Violin Plot 
similar to box plot. it shows the distribution of quantitative data across. the further it is farther from the center has the highest frequency. The higher the ball, the higher the people in that range

```python
plt.figure(figsize=(10,7))
plt.title('VIOLIN  PLOT')
sns.violinplot(x = 'day', y = 'total_bill', data = tips, palette = 'YlOrRd', hue = 'sex');
```

```python
plt.figure(figsize=(10,7))
plt.title('VIOLIN  PLOT')
sns.violinplot(x = 'day', y = 'total_bill', data = tips, palette = 'YlOrRd', hue = 'sex', split = True);
```

# Stipplot and Swarmplot
the stipplot will draw a scatter plot where one variable is catagorical.

a swarmplot is a stripplot but the opoint are adjusted so that they doesnt overlap. it does not scale well to 

```python
plt.figure(figsize=(10,7))
plt.title('VIOLIN  PLOT')
sns.stripplot(x = 'day', y = 'total_bill', data = tips, palette = 'Dark2', jitter = True,
               dodge = True, hue = 'sex');
```

```python
plt.figure(figsize=(10,7))
plt.title('VIOLIN  PLOT')
sns.swarmplot(x = 'day', y = 'total_bill', data = tips, palette = 'CMRmap',
               dodge = True, hue = 'sex');
```

```python
sns.violinplot(x = 'day', y = 'total_bill', data = tips, dodge = True);
sns.swarmplot(x = 'day', y = 'total_bill', data = tips,
               dodge = True, color = 'black');
```

# Catplot or Factorplot
This is the most general form of a categorical 

```python
sns.factorplot(x = 'sex', y = 'total_bill', data = tips,
               kind = 'bar', estimator = np.std);
```

```python
sns.factorplot(x = 'sex', y = 'total_bill', data = tips,
               kind = 'swarm', estimator = np.std);
```

# Rugplot
to check the distribution of a particular dataset using strip lines

no space btw the lines means there are many individual within that set. Focus on the density of the data


```python
sns.rugplot(tips['total_bill'], height = 0.3)
#to check the distribution of a series
```

# Jointplot
plots two different types of graph

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips);
#scatter plot shows you the near relationship btw the tip
```

# Distplot
It shows relationship btw a kde(kernel density estimation) and a histogram

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'], kde = False)
#bins can be use the increase and decrease the space within the distribution
```

# Pairplot
it is used to plot numerical columns against one another within a dataframe

```python
sns.pairplot(tips, hue = 'smoker', palette = 'Dark2')
```

# kde plot

```python
sns.kdeplot(tips.total_bill, color = 'green')
sns.rugplot(tips.total_bill, height = 0.6, color = 'red')
```

```python
sns.kdeplot(tips.total_bill, color = 'red')
sns.rugplot(tips.total_bill, height = 0.6, color = 'green')
```
